### 2. Gráfica de una función Coseno:


$$y(x)=\cos(x)$$

```python
# Ahora vamos a plotear una función coseno

# Generamos un eje X
x = np.arange(-2*np.pi, +2*np.pi, 0.01)

# Evaluamos el eje Y usando una función coseno
y = np.cos(x)
```

```python
# Graficamos:

plt.figure()

plt.plot(x,np.cos(x), label = "Funcion Coseno")

plt.grid()

plt.legend()

plt.show()
```

### 3. Gráfica de una célula:

Ahora vamos a descargarnos la imagen de una célula, tomada por un microscopio, y vamos a graficarla usando Python.


#### Repositorio de imágenes:

https://scikit-image.org/docs/stable/api/skimage.data.html


```python
# Importamos una librería de imágenes llamada scikit

import skimage

from skimage import data
```

```python
# Obtenemos la imagen de la célula

cell_image = data.cell()
```

```python
# Graficamos la imagen de la célula

fig, ax = plt.subplots(figsize=(6,6))

Z = ax.imshow(cell_image, cmap = "viridis")

plt.colorbar(Z)

plt.show()
```
